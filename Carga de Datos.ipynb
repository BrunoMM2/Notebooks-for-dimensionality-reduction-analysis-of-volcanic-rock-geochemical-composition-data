## Importamos los paquetes que necesitamos


```python
#importamos los paquetes que necesitamos
import pandas as pd
import numpy as np
```

## Definiremos los indices de las columnas

Definimos los

MAYORES = []
MENORES = []

en el formato en el que están en nuestro dataset del que los importaremos, nosotros también hemos añadido

FEOS = ['FE2O3(WT%)', 'FEO(WT%)']

ya que nuestro dataset no tiene en todas las muestras directamente el FEOT y habrá que calcularlo para algunas de ellas
 

```python
FEOS = ['FE2O3(WT%)', 'FEO(WT%)']
MAYORES = ['SIO2(WT%)', 'TIO2(WT%)','AL2O3(WT%)','FEOT(WT%)', 'CAO(WT%)', 'MGO(WT%)', 'MNO(WT%)', 'K2O(WT%)', 'NA2O(WT%)', 'P2O5(WT%)']
MENORES = ['LA(PPM)', 'CE(PPM)', 'PR(PPM)', 'ND(PPM)', 'SM(PPM)', 'EU(PPM)', 'GD(PPM)', 'TB(PPM)', 'DY(PPM)', 'HO(PPM)', 'ER(PPM)', 'YB(PPM)', 'LU(PPM)']
```

## Cargamos las columnas que nos interesen para nuestro analisis

Creamos el dataframe "tabla" leyendo con pandas los datos de nuestro csv y especificándole qué columnas queremos(de las definidas anteriormente)

```python
tabla = pd.read_csv('C:/Users/Bruno/Documents/IPNA/Departamento Volcanología/Clasificacion Rocas Volc UMAP/Datos/Hawaii/HAWAIIAN_ISLANDS_UNION funciona.csv',sep=";",usecols= MAYORES+FEOS)
```

```python
tabla.info()
```

```python
tabla
```

## Sustituimos NAN por 0

```python
df = tabla.replace(np.nan,0)
```

```python
df
```

## Actualizamos la columna de FEOT en base a FEO y FE2O3

Aplicamos las operaciones necesarias para realizar la conversion y al terminar eliminamos las columnas de FEO(WT%) y FE2O3(WT%)

```python
# Apply the logic using a lambda function
df['FEOT(WT%)'] = df.apply(lambda row: (row['FEO(WT%)'] / 0.77731) + ((row['FE2O3(WT%)'] * 0.69943)/ 0.77731) if row['FEOT(WT%)'] == 0 else row['FEOT(WT%)'], axis=1)
del df['FEO(WT%)']
del df['FE2O3(WT%)']
```

```python
df
```

## Filtrar las filas que no nos interesan

De cara filtrar podemos Crear una columna Total con la suma de los mayores y eliminar los que no lleguen a X%

```python
# Specify the columns to consider for the sum condition
#columns_to_sum = ['column1', 'column2']
columns_to_sum = MAYORES
# Specify the threshold value for the sum condition
threshold = 90

# Remove rows based on the sum condition
df = df[df[columns_to_sum].sum(axis=1) > threshold]
```

```python
df
```
